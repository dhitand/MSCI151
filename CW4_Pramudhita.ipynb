## Add a constant (intercept)

### Subtask:
Add a constant term to the features DataFrame for the intercept in the OLS model.

Add a constant term to the features DataFrame for the intercept in the OLS model.



```python
import statsmodels.api as sm

# Add a constant term to the features DataFrame for the intercept
X = sm.add_constant(X)

print("Constant term added to the features DataFrame.")
display(X.head())
```

## Fit the ols model with robust standard errors

### Subtask:
Use `statsmodels.api` to fit the OLS model with the specified features and target, and calculate robust standard errors.


**Reasoning**:
Use statsmodels.api to fit the OLS model with the specified features and target, and calculate robust standard errors.



```python
# Fit the OLS model
model = sm.OLS(y, X)
results = model.fit()

# Calculate robust standard errors (HC1)
robust_se = results.get_robust_cov_params(cov_type='HC1')

print("OLS model fitted and robust standard errors calculated.")
display(results.summary())
display(robust_se)
```

**Reasoning**:
The previous attempt to get robust standard errors failed because the method `get_robust_cov_params` was not found. I need to use the correct method to calculate robust standard errors, which is by specifying the `cov_type` argument directly in the `fit()` method.



```python
# Fit the OLS model with robust standard errors (HC1)
results = model.fit(cov_type='HC1')

print("OLS model fitted with robust standard errors.")
display(results.summary())
```

# Task
Perform Principal Component Analysis (PCA) on the dataset loaded from "/content/sample_data/MSCI151 Data.xlsx". The independent variables (X) for the PCA are 'debt', 'RnD', 'Firm Age', 'Firm Size', 'CEO Experience', 'foreign sales', 'CEO education', 'CEO Gender', 'marketing', 'Product Sentiment', and 'PoB'. The dependent variable (Y) is 'Profitability'. Provide the loading factors for each variable.

## Prepare the data

### Subtask:
Select the specified features (X) from the dataframe for PCA.


**Reasoning**:
Define the features and select them from the dataframe.



```python
features = ['debt', 'RnD', 'Firm Age', 'Firm Size', 'CEO Experience', 'foreign sales', 'CEO education', 'CEO Gender', 'Marketing', 'Product Sentimet', 'PoB' ]
X = df[features]

print("Features (X) selected for PCA.")
display(X.head())
```

## Standardize the data

### Subtask:
Standardize the feature data since PCA is affected by scale.


**Reasoning**:
Standardize the feature data for PCA.



```python
from sklearn.preprocessing import StandardScaler

# Instantiate a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the features DataFrame X
scaler.fit(X)

# Transform the features DataFrame X and store the standardized data
X_scaled = scaler.transform(X)

# Convert the scaled data back to a DataFrame for easier viewing (optional)
X_scaled = pd.DataFrame(X_scaled, columns=features)


print("Features (X) standardized.")
display(X_scaled.head())
```

## Perform pca

### Subtask:
Apply PCA to the standardized data stored in `X_scaled`.


**Reasoning**:
Import PCA, instantiate the model, fit it to the scaled data, and transform the scaled data.



```python
from sklearn.decomposition import PCA

# Instantiate a PCA object, keeping all components
pca = PCA(random_state=42)

# Fit PCA to the standardized data
pca.fit(X_scaled)

# Transform the standardized data
X_pca = pca.transform(X_scaled)

print("PCA applied to the standardized data.")
print(f"Shape of transformed data: {X_pca.shape}")
```

## Get loading factors

### Subtask:
Calculate and display the loading factors for each variable on the principal components.


**Reasoning**:
Access the principal components from the fitted PCA object and create a DataFrame to display the loading factors with appropriate labels.



```python
import pandas as pd

# Access the principal components (eigenvectors)
pca_components = pca.components_

# Create a DataFrame for better visualization of loading factors
# Rows are principal components, columns are original features
loading_factors = pd.DataFrame(pca_components.T, columns=[f'Principal Component {i+1}' for i in range(pca_components.shape[0])], index=features)

# Display the loading factors
print("Loading Factors for each variable on the Principal Components:")
display(loading_factors)
```

### Data Analysis Key Findings

*   The loading factors were successfully calculated and displayed, showing the contribution of each original variable to each principal component.
