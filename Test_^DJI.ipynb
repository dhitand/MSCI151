```python
import yfinance as yf
import pandas as pd

# Define the ticker symbol and date range
ticker_symbol = '^DJI'
start_date = '1997-01-01'
end_date = '2025-09-30'

# Download historical data
data = yf.download(ticker_symbol, start=start_date, end=end_date)

# Display the first few rows of the downloaded data
print("Downloaded data for ^DJI:")
display(data.head())
```

```python
# Calculate monthly returns
# Resample to monthly frequency, taking the last trading day's adjusted close price
monthly_prices = data['Close'].resample('ME').ffill()

# Calculate percentage change for monthly returns
monthly_returns = monthly_prices.pct_change().dropna()

print("Monthly Returns for ^DJI:")
display(monthly_returns.head())
```

```python
# Save the monthly returns to an Excel file
excel_file_name = 'DJI_Monthly_Returns.xlsx'
monthly_returns.to_excel(excel_file_name)

print(f"Monthly returns saved to {excel_file_name}")
```

You can download the file `DJI_Monthly_Returns.xlsx` from the file browser in the left sidebar.
