```python
import pandas as pd

file_path = '/content/sample_data/Global Policy Uncertainty Data with US.xlsx'
df = pd.read_excel(file_path)
print(df.head())
```

```python
import statsmodels.api as sm

# Define the dependent variable (Y) and independent variable (X)
Y = df['US']
X = df['GEPU']

# Add a constant (intercept) to the independent variable
X = sm.add_constant(X)

# Create and fit the OLS regression model
model = sm.OLS(Y, X)
results = model.fit()

# Print the regression summary
print(results.summary())
```

# In conclusion, EPU doesn't affect the US Stock Market since looking the R-squared, Adj. R-squared, Prob (F-statistic) more than 0,1, and P>|t| more than 0,1
