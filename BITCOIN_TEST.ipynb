```python
import yfinance as yf

tickers = ["BTC-USD", "GC=F", "^DJI", "CL=F"]
start_date = "2025-01-01"
end_date = "2025-06-30"

data = yf.download(tickers, start=start_date, end=end_date)

display(data.head())
```

```python
# Select 'Close' prices and drop rows with any NaN values
close_prices = data['Close'].dropna()

# Calculate daily returns
returns = close_prices.pct_change().dropna()

# Define the new names for the tickers
new_names = {
    'BTC-USD': 'bitcoin',
    'GC=F': 'gold',
    '^DJI': 'US',
    'CL=F': 'oil'
}

# Rename the columns
returns_renamed = returns.rename(columns=new_names)

display(returns_renamed.head())
```

```python
import statsmodels.api as sm

# Define the dependent variable (Y)
Y = returns_renamed['bitcoin']

# Define the independent variables (X)
X = returns_renamed[['gold', 'oil', 'US']]

# Add a constant (intercept) to the independent variables
X = sm.add_constant(X)

# Create and fit the OLS model
model = sm.OLS(Y, X)
results = model.fit()

# Print the regression summary
print(results.summary())
```
